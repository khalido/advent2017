# Part 2

Whats the first repeated frequency?

This is straightforward - I'm keeping a set of every frequency and printing the first repeated one:

```python
freq = 0
freqs_set = set([freq])
freqs_list = [freq]

for f in cycle(inp):
    freq += f
    freqs_list.append(freq)
    if freq in freqs_set:
        print(f"{freq} is the repeating frequency")
        break
    else:
        freqs_set.add(freq)
```

part 2 ans: `56360`

Now to eyeball it:

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.set_title("Freqency as it changes")
ax.set_ylabel("Frequency")
ax.plot(freqs_list);
```
