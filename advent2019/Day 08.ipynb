```python
test_inp = "123456789012"

def parse_input(inp, width=3, height=2):
    inp = [int(i) for i in inp]
    length = len(inp) // (width * height)
    return np.reshape(inp, (length, height, width))

inp = parse_input(test_inp)
inp
```

```python
def solve_1(raw_inp=raw_inp):
    
    total_nums = 25 * 6
    idx, zero_count = 0, 0
    layers = parse_input(raw_inp, 25, 6)

    for i, layer in enumerate(layers):
        non_zeros = np.count_nonzero(layer)
        if non_zeros > zero_count:
            zero_count = non_zeros
            idx = i
        
        clear_output(wait=True)
        fig, ax = plt.subplots(figsize=(8,4))
        ax.imshow(layer)
        ax.set_title(f"Processing layer {i}")
        plt.show()
        sleep(0.03)
        

    one_count = np.count_nonzero(layers[idx] == 1)
    two_count = np.count_nonzero(layers[idx] == 2)
    print(f"Ones: {one_count}, Twos: {two_count} | Times: {one_count * two_count}")
    return one_count * two_count

solve_1()
```

```python
colours = {0: "black", 
           1: "white",
           2: "transparent"}
```

```python
cmap = colors.ListedColormap(['#66000000','white','transparent'])
bounds=[-0.5,0.5,1.5, 2.5]
norm = colors.BoundaryNorm(bounds, cmap.N)
```

```python
test_inp = "0222112222120000"
inp = parse_input(test_inp, 2, 2)
```

```python
def get_pixel(layers, pos):
    """takes in an array of images and pos and returns the first non empty pixel"""
    
    for layer in layers[1:]:
        pixel = layer[pos]
        if pixel != 2:
            return pixel
    print("Shouldn't have gotten here")
```

```python
def solve_2(raw_inp=raw_inp, width=25, height=6):
    
    inp = parse_input(raw_inp, 25, 6)
    
    # start with output being same as first layer 
    im = np.copy(inp[0])
    # and set all the transparent pixels to -99 for processing
    #im[im == 2] = -99

    for i in np.arange(height):
        for j in np.arange(width):
            pos = (i, j)
            pixel = im[pos]

            if pixel == 2:
                im[pos] = get_pixel(inp, pos)
        
        # return if no more transparent pixels
        if np.count_nonzero(im == 2) == 0:
            print(f"Finished! at {pos}")
            return im
        
    return im

im = solve_2()
plt.imshow(im);
```
