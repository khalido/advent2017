```python
import utils
import math
from itertools import combinations
import matplotlib.pyplot as plt
import plotly.express as px
```

## Day 1: Report Repair

[#](https://adventofcode.com/2020/day/1) We have a list of numbers and need to find the two numbers that add up to 2020, and multiply them to get the answer.

```python
test1 = """1721
979
366
299
675
1456""".splitlines()

inp1 = utils.get_input(1)

def parse(inp):
    return [int(i) for i in inp]

parse(test1)
```

```python
def solve_1(inp, numbers=2):
    # get the first nums which add up to 2020
    nums = [i for i in combinations(parse(inp), numbers) if sum(i)==2020][0]
    return math.prod(nums)

assert solve_1(test1) == 514579
solve_1(inp1)
```

To eyeball the numbers:

```python
lx = list(range(250,2020))
ly = list(range(2020-250, 0,-1))
lx[:10], ly[:10]
```

```python
# the points to scatter
nums = [i for i in combinations(parse(inp1), 2)]
x, y = zip(*nums)

# the winning point
X, Y = [i for i in nums if sum(i)==2020][0]

# the bingo line
lx = list(range(250,2020))
ly = list(range(2020-250, 0,-1))

fig = plt.figure(figsize=(10,6))
plt.title(f"Number pairs - Winner: {X, Y}")
plt.scatter(x, y, alpha=0.4, s=3)
plt.scatter(X, Y, color="red", s=140, alpha=0.5)
plt.plot(lx, ly, lw=5, alpha=0.2, color="green");
```

**Part 2:** Now we want to find the three numbers which sum to 2020 and return the multiple.

Since I made a general func to solve for 1, this is just rerunning the above with numbers set to 3.

```python
assert solve_1(test1, 3) == 241861950
solve_1(inp1, 3)
```

```python
nums = [i for i in combinations(parse(inp1), 3)]
x, y, z = zip(*nums)
X, Y, Z = [i for i in nums if sum(i)==2020][0]

from mpl_toolkits.mplot3d import Axes3D 

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
plt.title("Number pairs")

ax.scatter(x,y, z, alpha=0.2, s=2.5)
ax.scatter(X, Y, Z, color="red", s=140, alpha=0.6);
```
